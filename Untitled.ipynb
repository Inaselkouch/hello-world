```python
import numpy as np
import pandas as pd
d_gender=pd.read_csv("Person_Gender_Height_Weight_Index.csv")
d_gender.head()
```

```python
#creare nuovo dataset con tutte le colonne tranne Index
bmi_ds=pd.DataFrame(d_gender, columns=['Gender', 'Height','Weight'])
bmi_ds
```

```python
#salvare il data frame bmi_ds come dataset csv in locale
bmi_ds.to_csv('es_gender.csv', index=False)
#ora visualizzare il csv normalmente
d1_gender=pd.read_csv("es_gender.csv")
d1_gender.head()
```

```python
#aggiungere colonna BMI che contiene l'operazione PESO DIVISO ALTEZZA AL QUADRATO
#prima cosa elevo l'altezza al quadrato
d1_gender["Height^2"]=(d1_gender["Height"]**2)
d1_gender["Height^2"]
```

```python
#seconda cosa divido
d1_gender["Weight/Height**2"] = (d1_gender["Weight"] / d1_gender["Height^2"])
d1_gender["Weight/Height**2"]
```

```python
#adesso posso nominare la nuova colonna BMI 
BMI=d1_gender["Weight/Height**2"]
BMI
```

```python
#ora aggiungo la colonna nel data
Tab_finale=pd.DataFrame(d1_gender, columns=['Gender', 'Height','Weight','BMI'])
Tab_finale
```
